```python
import mouse_acceleration_models as mam

zoom_max_x = 50

acc_factor_base = 1
acc_factor_max = 2.2
acc_factor_calc = acc_factor_max - acc_factor_base
acc_rate = 0.9

first_rise = 0.1

rise_offset = mam.sigmoid_inverse(first_rise, acc_factor_calc, acc_rate)
rise_offset

print(f'First rise y value: {first_rise}; offset: {rise_offset}')

# rise_offset = -2
offset = rise_offset - 1 - 10

df = mam.display_curve(
    'sigmoid_pure',
    acc_factor_base=acc_factor_base,
    acc_factor_max=acc_factor_max,
    acc_rate=acc_rate,
    max_speed=0,
    start_offset=offset,
    zoom_max_x=zoom_max_x,
    should_round=False,
    range_start=-10,
    # range_end=50,
)

df.head(zoom_max_x)
```
