# Acceleration Curve Previews

## Sigmoid Curve

```python
import mouse_acceleration_models as mam

base_sensitivity = 0.5
max_sensitivity = 2.5
accel_slope = 1
max_speed = 120
offset = 5

df = mam.gen_sensitivity_curve_sigmoid_df(
    0,
    254,
    base_sensitivity,
    max_sensitivity,
    accel_slope,
    max_speed=max_speed,
    offset=offset,
    should_round=True,
)

df_zoomed = df.head(30)

mam.plot_sensitivity(
    df_zoomed['x'], df_zoomed['sensitivity'], offset=offset,
    title='Sensitivity (Zoomed)',
)
mam.plot_output_speed(
    df_zoomed['x'], df_zoomed['output'], offset=offset,
    title='Output Speed (Zoomed)',
)

mam.plot_sensitivity(df['x'], df['sensitivity'], offset=offset)
mam.plot_output_speed(df['x'], df['output'], offset=offset)

display(df)
```
